```python
import seaborn as sns
%matplotlib inline

```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
grids = sns.PairGrid(iris)
grids
```

```python
#grids.map(plt.scatter)
grids.map_diag(sns.displot)
```

```python
import pandas as pd; 
#read the data from the xlsx file
data = pd.read_excel('Copy_of_UNC_Weight_Management_Feedback_March.25.2022_00.50.xlsx')
```

```python
# get data head
data.head()
```

```python
# remove the second row because its repeating the headers
data1 = data.drop(data.index[0])
#diplay the data header without the second row
data1.head()
```

```python
# remove rows with blank Duration (in seconds)
data2 = data1.dropna(subset=['Duration (in seconds)'])
```

```python
# import stats which will use to compute the t-test
import scipy.stats as stats

# get the duration column head
duration_column = data2['Duration (in seconds)']
duration_column.head()
```

```python
# calculate ttest using stats.ttest_1samp of duration_column since its a one sample t-test
t_test_duration = stats.ttest_1samp(duration_column, duration_column.mean())
```

```python
# print the result with 4 decimal
print('t-test for duration: ', round(t_test_duration.statistic, 4), ' and p-value: ', round(t_test_duration.pvalue, 4))
```

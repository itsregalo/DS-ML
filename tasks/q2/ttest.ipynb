```python
import pandas as pd; 
import numpy as np; 
import matplotlib.pyplot as plt
```

```python
#read the data from the xlsx file
df = pd.read_excel('Copy_of_UNC_Weight_Management_Feedback_March.25.2022_00.50.xlsx')
```

```python
df.head()
```

```python
# remove the second row
df1 = df.drop(df.index[0])

df1.head()
```

```python
# t -test for the mean of the Duration (in seconds)
duration_mean = df1['Duration (in seconds)'].mean()
# independent sample t-test of duration_mean
from scipy import stats
t_test_duration = df1['Duration (in seconds)'].mean()
t_test_duration_std = df1['Duration (in seconds)'].std()
t_test_duration_std_div_sqrt = t_test_duration_std/np.sqrt(len(df1['Duration (in seconds)']))
t_test_duration_t_stat = (duration_mean - df1['Duration (in seconds)'].mean())/t_test_duration_std_div_sqrt
t_test_duration_p_value = 1 - stats.t.cdf(t_test_duration_t_stat, len(df1['Duration (in seconds)']))

# print the results
print('t-test for the mean of the Duration (in seconds)')
print('t-statistic:', t_test_duration_t_stat)
print('p-value:', t_test_duration_p_value)

```

```python
# scipy t-test for the mean of the Duration (in seconds)
from scipy import stats
t_test_duration = stats.ttest_1samp(df1['Duration (in seconds)'], df1['Duration (in seconds)'].mean())

# print the results
print('t-test for the mean of the Duration (in seconds)')
print('t-statistic:', t_test_duration[0])
print('p-value:', t_test_duration[1])
```
